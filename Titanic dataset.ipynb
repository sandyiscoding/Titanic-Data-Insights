Task 1: How many survived classwise? Display percentage.

```python
df_pclass_survival = df[['Pclass','Survived']].groupby(['Pclass']).mean().reset_index()
df_pclass_survival[['Survived']]=df_pclass_survival[['Survived']]*100
df_pclass_survival.rename(columns={'Survived':'Survival in %'},inplace=True)
```

```python
sns.barplot(x='Pclass',y='Survival in %',data=df_pclass_survival)
plt.show()
```

Task 2: How many survived Genderwise? Display percentage.

```python
df_gender_survival = df[['Sex','Survived']].groupby('Sex').mean().reset_index()
df_gender_survival[['Survived']]=df_gender_survival[['Survived']]*100
df_gender_survival.rename(columns={'Survived':'Survival in %'},inplace=True)
```

```python
sns.barplot(x='Sex',y='Survival in %',data=df_gender_survival)
plt.show()
```

Task 3: Extract the salutation of each person and give the count

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.')
```

```python
df
```

```python
df['Title'].unique()
```

```python
df['Title'].value_counts()
```

```python
def replace_title(row):
  title = row['Title']
  relevant_titles = ['Mr', 'Mrs', 'Miss', 'Master']
  if title in relevant_titles:
    return title
  else:
    return "Others"
```

```python
df['Title'] = df.apply(replace_title, axis=1)
```

```python
df
```

```python
df['Title'].unique()
```
